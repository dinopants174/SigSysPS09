```python
from __future__ import print_function, division
import sys
import numpy as np
import matplotlib.pyplot as mplib
import scipy.interpolate as interp
from matplotlib.widgets import Slider

# based on an example from Joe Kington
# this is a quantized slider which increments in quantized steps that are stepsize wide
# the default stepsize is 1 which means that the slider by default will go through all the integers
class QuantSlider(Slider):
    def __init__(self, *args, **kwargs):
        self.step = kwargs.pop('stepsize', 1)
        Slider.__init__(self, *args, **kwargs)

    def set_val(self, x):
        quant = int(x / self.step) * self.step+self.valinit-self.step
        xy = self.poly.xy
        xy[2] = quant, 1
        xy[3] = quant, 0
        self.poly.xy = xy
        self.valtext.set_text(self.valfmt % quant)
        if self.drawon: 
            self.ax.figure.canvas.draw()
        self.val = x
        if not self.eventson: 
            return
        for cid, func in self.observers.iteritems():
            func(quant)
```

The following functions implement 3 different interpolation techniques to generate a continuous-time signal from a discrete-time signal

```python
# this function implements a hold interpolation of a discrete-time sequence
# it uses the built-in interp1d function from  scipy.signal 

def hold_interp(x_dt, dt_times, ct_times):
    x_lin_interp = interp.interp1d(dt_times, x_dt, kind='zero')
    x = x_lin_interp(ct_times) 
    return x

def linear_interp(x_dt, dt_times, ct_times):
    x_lin_interp = interp.interp1d(dt_times, x_dt, kind='linear')
    x = x_lin_interp(ct_times) 
    return x

def sinc_interp(x_dt, dt_times, ct_times, plot):
    x = np.zeros(len(ct_times))
    samp_period = dt_times[1]-dt_times[0]
   # for n in range(0, len(x)):
    for k in range(0, len(x_dt)):
        x = x+x_dt[k]*np.sinc(1/samp_period*(ct_times-dt_times[k]))
        if plot==True:
            mplib.plot(ct_times,x_dt[k]*np.sinc(1/samp_period*(ct_times-dt_times[k])), color='0.75')    
    return x
```

The following cell produces a figure which illustrates  three different interpolation methods. You can move the slider to visualize what different sampling periods look like. Ts is the symbol period here.

```python
# these are the times we shall evaluate the continuous-time wave form for
ts = np.linspace(0,1, 512)
# create a generic waveform by interpolating a sequence of numbers using cubic interpolation
# this will serve as our approximation for a continuous signal
generic_ct_signal = interp.interp1d(np.array([0,0.1,0.4, 0.7, 1]), np.array([0.1,1.0,0.3, 1,0.1]), kind='quadratic')

# create a plot 
# we are going to need 2 axes, one for the figure
# and another for the slider
fig, ax = mplib.subplots()
# set the axis range for the main plot
 

# this is the function that gets called whenever the slider is changed
# the value of the slider is passed as M
def on_change(Ts):
    sampling_times = np.linspace(0,1, np.round(1/Ts))
    
    # refresh the line
    mplib.subplot(4,2,1)
    mplib.cla()
    mplib.stem(sampling_times, generic_ct_signal(sampling_times), linefmt='b-', markerfmt='ro', basefmt='b-')
    mplib.ylabel('DT Samples')    
    mplib.axis([0,1, -0.5, 2])    
    
    # compute the continuous reconstructions using the hold, linear and sinc interpolation
    x_recon_hold = hold_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts) 
    x_recon_lin = linear_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts) 
 
    mplib.subplot(4,2,3)
    mplib.cla()
    mplib.plot(ts, x_recon_hold, 'b')  
    mplib.plot(sampling_times, generic_ct_signal(sampling_times), 'or')
    
    #plot the reconstructed signal using hold interpolation   
    mplib.ylabel('Hold')    
    mplib.axis([0,1, -0.5, 2])
    # plot the spectrum of the reconstruced continuous time signal
    mplib.subplot(4,2,4)
    mplib.cla()
    # first create the x-axis index values
    centered_fft_indices = np.arange(-int(len(x_recon_lin)/2),int(len(x_recon_hold)/2),1)
    # plot the absolute value of the FFT of the time domain signal on a dB scale
    # fftshift moves the elements of the fft around to be centered around frequncy zero       
    mplib.plot(centered_fft_indices,20*np.log10(np.fft.fftshift(np.abs(np.fft.fft(x_recon_hold)))))
    mplib.axis([-300, 300, -100, 60])
    mplib.grid()
    mplib.ylabel('dB')

    #plot the reconstructed signal using linear interpolation
    mplib.subplot(4,2,5)
    mplib.cla()
    mplib.plot(ts, x_recon_lin, 'b')  
    mplib.plot(sampling_times, generic_ct_signal(sampling_times), 'or')
    mplib.ylabel('Linear')    
    mplib.axis([0,1, -0.5, 2])
    
    # plot the spectrum of the reconstruced continuous time signal
    mplib.subplot(4,2,6)
    mplib.cla()
    # first create the x-axis index values
    centered_fft_indices = np.arange(-int(len(x_recon_lin)/2),int(len(x_recon_lin)/2),1)
    # plot the absolute value of the FFT of the time domain signal on a dB scale
    # fftshift moves the elements of the fft around to be centered around frequncy zero    
    mplib.plot(centered_fft_indices,20*np.log10(np.fft.fftshift(np.abs(np.fft.fft(x_recon_lin)))))
    mplib.axis([-300, 300, -100, 60])
    mplib.grid()
    mplib.ylabel('dB')
    
    #plot the reconstructed signal using the sinc interpolation
    mplib.subplot(4,2,7)
    mplib.cla()
    # compute the reconstruction of the sinc signal, include a plot of the underlying sinc functions
    x_recon_sinc = sinc_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, plot=True) 
    mplib.plot(ts, x_recon_sinc, 'b')  
    # add the DT samples that were used
    mplib.plot(sampling_times, generic_ct_signal(sampling_times), 'or')
    mplib.ylabel('Sinc')    
    mplib.axis([0,1, -0.5, 2])
    
    # plot the spectrum of the reconstruced continuous time signal
    mplib.subplot(4,2,8)
    mplib.cla()
    # first create the x-axis index values
    centered_fft_indices = np.arange(-int(len(x_recon_sinc)/2),int(len(x_recon_sinc)/2),1)
    # plot the absolute value of the FFT of the time domain signal on a dB scale
    # fftshift moves the elements of the fft around to be centered around frequncy zero            
    line1,=mplib.plot(centered_fft_indices,20*np.log10(np.abs(np.fft.fftshift(np.fft.fft(x_recon_sinc)))), label='Sinc Interp.')
    mplib.axis([-300, 300, -100, 60])
    mplib.grid()
    mplib.ylabel('dB')

    
    # redraw to update
    mplib.draw()

# run the figure update function once for the first time with Ts = 0.07    
on_change(0.07)    

fig.text(0.65,0.95,'Frequency Domain', fontsize=20)
fig.text(0.25,0.95,'Time Domain', fontsize=20)    
# make axes for the slider
slider_ax = mplib.axes([0.3, 0.02, 0.6, 0.02])
# make a  slider that will do stepsizes of 2 with minimum of 3 and maximum of 65
slider = Slider(slider_ax, "Ts", 0.01, 0.1,  valinit=0.07, valfmt=u'%.2f')
# set the function that is called when slider is changed
slider.on_changed(on_change)
# display the figure
mplib.show()    
```

```python
ts = np.linspace(0,1, 512)
# create a generic waveform by interpolating a sequence of numbers using cubic interpolation
# this will serve as our approximation for a continuous signal
generic_ct_signal = interp.interp1d(np.array([0,0.1,0.4, 0.7, 1]), np.array([0.1,1.0,0.3, 1,0.1]), kind='quadratic')

# create a plot 
# we are going to need 2 axes, one for the figure
# and another for the slider
fig, ax = mplib.subplots()
# set the axis range for the main plot
 

# this is the function that gets called whenever the slider is changed
# the value of the slider is passed as M
def on_change_spec(Ts):
    mplib.subplot(1,1,1)
    sampling_times = np.linspace(0,1, np.round(1/Ts))
    
    # compute the continuous reconstructions using the hold, linear and sinc interpolation
    x_recon_hold = hold_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts) 
    x_recon_lin = linear_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts) 
    x_recon_sinc = sinc_interp(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, plot=False) 
    
    # plot the spectrum of the reconstruced continuous time signal
    mplib.cla()
    # first create the x-axis index values
    centered_fft_indices = np.arange(-int(len(x_recon_sinc)/2),int(len(x_recon_sinc)/2),1)
    # plot the absolute value of the FFT of the time domain signal on a dB scale
    # fftshift moves the elements of the fft around to be centered around frequncy zero            
    line3Spec,=mplib.plot(centered_fft_indices,20*np.log10(np.fft.fftshift(np.abs(np.fft.fft(x_recon_hold)))), 'm', label='Hold Interp.')
    line2Spec,=mplib.plot(centered_fft_indices,20*np.log10(np.fft.fftshift(np.abs(np.fft.fft(x_recon_lin)))), 'c', label='Linear Interp.')
    line1Spec,=mplib.plot(centered_fft_indices,20*np.log10(np.abs(np.fft.fftshift(np.fft.fft(x_recon_sinc)))), label='Sinc Interp.')
    mplib.legend(handles=[line3Spec, line2Spec, line1Spec])
    mplib.axis([-300, 300, -100, 60])
    mplib.grid()
    mplib.ylabel('Magnitude of Spectrum (dB)')
    mplib.draw()

    
    # redraw to update
    mplib.draw()

on_change_spec(0.07)    
# make axes for the slider
slider_ax = mplib.axes([0.3, 0.02, 0.6, 0.02])
# make a  slider that will do stepsizes of 2 with minimum of 3 and maximum of 65
slider = Slider(slider_ax, "Ts", 0.01, 0.1,  valinit=0.07, valfmt=u'%.2f')
# set the function that is called when slider is changed
slider.on_changed(on_change_spec)
# display the figure
mplib.show()  
```

The cell below helps illustrate how the sinc interpolation works. You will need to close all other figures

```python
from matplotlib.widgets import Button

ts = np.linspace(0,1, 512)
# create a generic waveform by interpolating a sequence of numbers using cubic interpolation
# this will serve as our approximation for a continuous signal
generic_ct_signal = interp.interp1d(np.array([0,0.1,0.4, 0.7, 1]), np.array([0.1,1.0,0.3, 1,0.1]), kind='quadratic')

# pick a nominal sampling period
Ts = 0.03
sampling_times = np.linspace(0,1, np.round(1/Ts))

# create axes
fig, ax = mplib.subplots()
mplib.subplots_adjust(bottom=0.2)


# this function helps animate a sinc interpolation
def sinc_interp_anim(x_dt, dt_times, ct_times, num_plots, plot_sum = False):
    x = np.zeros(len(ct_times))
    samp_period = dt_times[1]-dt_times[0]
    if num_plots > len(x_dt):
        num_plots = len(x_dt)
    if num_plots < 0:
        numplots = 0
    # plot num_plots sincs centered on each sample  value
    # and scaled by the height of the sample values
    for k in range(0, num_plots):
        x = x+x_dt[k]*np.sinc(1/samp_period*(ct_times-dt_times[k]))
        mplib.plot(ct_times,x_dt[k]*np.sinc(1/samp_period*(ct_times-dt_times[k])), 'm', lw = 1.5) 
        
    mplib.stem(dt_times, x_dt, linefmt='b-', markerfmt='ro', basefmt='b-')
    # if we need to plot the sum of the sincs, do it
    if plot_sum == True:
        mplib.plot(ct_times, x, 'k--', lw = 2)
    

# this class handles the plotting operations in response to button presses
class PlotSincInterp:
    
    NumSincs = 0
    plot_sum = False

    # this method increments the number of sincs plotted and calls the animation function
    def next(self, event):
        self.NumSincs += 1
        # wrap around
        if self.NumSincs > len(sampling_times):
            self.NumSincs = len(sampling_times)
        mplib.subplot(111)
        mplib.cla()
        sinc_interp_anim(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, num_plots = self.NumSincs, plot_sum = self.plot_sum) 
        mplib.draw()
    # this method decrements the number of sincs plotted and calls the animation function   
    def prev(self, event):
        # wrap around
        self.NumSincs -= 1
        if self.NumSincs < 0:
            self.NumSincs = 0
        mplib.subplot(111)
        mplib.cla()
        sinc_interp_anim(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, num_plots = self.NumSincs, plot_sum = self.plot_sum) 
        mplib.draw()
    # this method toggles displaying the sum of the sincs
    def sum(self, event):
        mplib.subplot(111)
        mplib.cla()
        self.plot_sum = not self.plot_sum
        sinc_interp_anim(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, num_plots = self.NumSincs, plot_sum = self.plot_sum)  
        mplib.draw()  

# call the animation function once to make the initial plot
sinc_interp_anim(x_dt = generic_ct_signal(sampling_times), dt_times = sampling_times, ct_times = ts, num_plots = 0) 

# this object handles the button presses
callback = PlotSincInterp()

# make the buttons
axprev = mplib.axes([0.7, 0.05, 0.1, 0.075])
axnext = mplib.axes([0.81, 0.05, 0.1, 0.075])
axsum = mplib.axes([0.59, 0.05, 0.1, 0.075])
bnext = Button(axnext, 'One More')
bnext.on_clicked(callback.next)
bprev = Button(axprev, 'One Less')
bprev.on_clicked(callback.prev)

bsum = Button(axsum, 'Toggle Sum')
bsum.on_clicked(callback.sum)

mplib.show()
```
